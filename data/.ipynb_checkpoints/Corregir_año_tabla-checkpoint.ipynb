```python
import pandas as pd
import numpy as np
import datetime
```

```python
df = pd.read_csv('nfl_football_team_stats.csv', index_col=0)

# Show dataframe
df.head()
```

```python
#ver el tipo de dato que es Date
df.dtypes
```

```python
#no existen cambos nulos
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date'].dtype
```

```python
df['Date'] = df['Date'].astype('datetime64[ns]')
```

```python
df['Year_correct'] = df['Date'].dt.year 
df['Month'] = df['Date'].dt.month 
```
